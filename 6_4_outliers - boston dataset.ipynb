```python
#Import the libraries
import numpy as np
import pandas as pd
from sklearn.datasets import load_boston,load_iris
import scipy as sc
import seaborn as sns
%matplotlib inline
```

```python
#Load the data
boston = load_boston()

#Find features and target
x = boston.data
y = boston.target
```

```python
print(boston.DESCR)
```

```python
#Create dataframe
boston_df = pd.DataFrame(boston.data)
boston_df.columns = boston.feature_names
boston_df['MEDV'] = boston.target

boston_df_o = boston_df.copy()
boston_df.shape
```

```python
boston_df
```

```python
sns.jointplot(x='RM', y='MEDV', data=boston_df)
```

```python
r = sc.stats.linregress(boston_df['RM'], boston_df['MEDV']).rvalue
```

```python
print(f'The measure R^2 is {r*r}')
```

# Z score

![](z_scores.png)


```python
from scipy import stats
import numpy as np
z = np.abs(stats.zscore(boston_df[['RM', 'MEDV']]))
print(z)
```

```python
boston_df_o = boston_df[(z < 3).all(axis=1)]
boston_df_o.shape
```

```python
boston_df.shape
```

```python
sns.jointplot(x='RM', y='MEDV', data=boston_df_o)
```

```python
r = sc.stats.linregress(boston_df_o['RM'], boston_df_o['MEDV']).rvalue
print(f'The measure R^2 is {r*r}')
```

# IQR based outliers

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.boxplot(x=boston_df['RM'])
```

```python
boston_df_o1 = boston_df.copy()

Q1 = boston_df_o1['RM'].quantile(0.25)
Q3 = boston_df_o1['RM'].quantile(0.75)
IQR = Q3 - Q1
print(IQR)
threshold_IQR = 1.5
boston_df_out = boston_df_o1[boston_df_o1['RM'] > (Q1 - threshold_IQR * IQR) ]
```

```python
boston_df_out.shape
```

```python

boston_df_out = boston_df_out[boston_df_out['RM'] < (Q3 + threshold_IQR * IQR)]
```

```python
boston_df_out.shape
```

```python
sns.boxplot(x=boston_df_out['RM'])
```

```python
sns.jointplot(x='RM', y='MEDV', data=boston_df_out)
```

```python
r = sc.stats.linregress(boston_df_out['RM'], boston_df_out['MEDV']).rvalue
print(f'The measure R^2 is {r*r}')
```
